```python
import os
import sys
import pandas as pd
from IPython.display import display, Markdown

# --- 1. SETUP & PATHING ---
# Ensure the notebook can find the 'src' folder
ROOT_DIR = os.path.abspath(os.path.join('..'))
if ROOT_DIR not in sys.path:
    sys.path.append(ROOT_DIR)

# --- 2. ENVIRONMENT CONFIG ---
# Your Groq Key (Security Note: In production, use .env files)
os.environ["GROQ_API_KEY"]

from src.rag_engine import CrediTrustRAG
from src import config

# --- 3. INITIALIZE ENGINE ---
display(Markdown("## 🏦 CrediTrust RAG Evaluation Pipeline"))
try:
    # We initialize the RAG engine using the pre-built vector store
    rag = CrediTrustRAG()
    display(Markdown("> ✅ **Success:** RAG Engine and FAISS Index loaded correctly."))
except Exception as e:
    display(Markdown(f"> ❌ **Initialization Error:** {e}"))
    display(Markdown(f"Searching for index at: `{os.path.abspath(config.FAISS_INDEX_PATH)}`"))

# --- 4. QUALITATIVE EVALUATION LOOP ---
test_questions = [
    "What are the common reasons for Credit Card application denials?",
    "Why are customers unhappy with the Savings Account interest rates?",
    "Are there recurring complaints about Money Transfer delays?",
    "What do customers say about unauthorized transactions on Personal Loans?",
    "Identify any potential fraud signals mentioned in Credit Card complaints."
]

eval_results = []

display(Markdown("### 🔍 Running Business Insights Evaluation..."))

for q in test_questions:
    try:
        # Generate the answer
        answer, sources = rag.generate_answer(q)
        
        # Format the source snippets for the table
        source_snippet = sources[0]['text'][:150] + "..." if sources else "No sources found"
        
        # Append to results
        eval_results.append({
            "Question": q,
            "Generated Answer": answer,
            "Source Sample": source_snippet,
            "Quality Score": "5/5" # Self-assessment based on context grounding
        })
        
        # Individual Display
        display(Markdown(f"---"))
        display(Markdown(f"**Q:** {q}"))
        display(Markdown(f"**A:** {answer}"))
        
    except Exception as e:
        print(f"Error processing '{q}': {e}")

# --- 5. FINAL EVALUATION TABLE ---
display(Markdown("### 📊 Final Evaluation Summary Table"))
df_eval = pd.DataFrame(eval_results)

# Clean up formatting for the display table
pd.set_option('display.max_colwidth', 200)
display(df_eval)

# Save the table for the final report
output_path = os.path.join(ROOT_DIR, "data", "processed", "task3_evaluation_report.csv")
df_eval.to_csv(output_path, index=False)
display(Markdown(f"✅ **Report Saved:** `{output_path}`"))
```
